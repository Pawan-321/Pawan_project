```python
import plotly.express as px
import numpy as np

# Create sample data
data = np.random.normal(30, 10, 1000)  # 1000 points with mean 30 and std 10

# Create histogram
fig = px.histogram(
    x=data,
    nbins=30,  # number of bins
    title='Sample Distribution',
    labels={'x': 'Value', 'y': 'Frequency'},
)

# Improve layout
fig.update_layout(
    plot_bgcolor="White",
    bargap=0.1,  # gap between bars
)

fig.show()
```
