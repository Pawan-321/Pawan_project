```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Collect data (replace with your actual input)
data = []
while len(data) < 5:  # Collect at least five numbers
    try:
        num = int(input("Enter a number: "))
        data.append(num)
    except ValueError:
        print("\nPlease enter valid integers.")
```

```python
# Calculate histogram
counts, bins = np.histogram(data, bins=5)
```

```python
# Plot histogram
plt.figure(figsize=(10, 6))
plt.bar(bins[:-1], counts, width=np.diff(bins), align='edge')
plt.title('Histogram of Data')
plt.xlabel('Value Range')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
